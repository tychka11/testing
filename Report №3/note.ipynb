- для чого ці ассерти, акщо ми можемо наприкла вказувати що саме ми очікуємо отримати, наприклад у функціях
- в пайтоні є нестрога типізація даних, на відміну від Java, GO, C

```python
def accept_int(i: int) -> int:
    """Нестрого типізована функція
    """
    i+= 12
    return i

print(accept_int(1.5))

def force_accept_only_int(i: int) -> int:
    """Строга типізація, функція впаде якщо не передати int"""
    # assert isinstance(i, int), f"У функцію було нередона {type(i)} а має бути {int}"
    i+= 12
    return i

print(force_accept_only_int(1.5))
```

## порівняння Assert та простої перевірки

- якщо ми перевіряємо умови і в нас true false - то чи можемо ми просто використати if
- такі перевірки мають право на існування, наприклад у Варнінгах, коли ми поперджаємо що щось може працювати або і непрацювати (негарантовано)

```python
version_restirciton = [3.12, 3.13, 3.10]
current_version = 3.9
if current_version not in version_restirciton:
    print("WARN: Ця версія несумісна і може призвести до помилок")

# assert current_version in version_restirciton, f"Версія {current_version} непідтримується"
```

```python
---------------------------------------------------------------------------
AssertionError                            Traceback (most recent call last)
Cell In[35], line 6
      3 if current_version not in version_restirciton:
      4     print("WARN: Ця версія несумісна і може призвести до помилок")
----> 6 assert current_version in version_restirciton, f"Версія {current_version} непідтримується"

AssertionError: Версія 3.9 непідтримується
```

- ми можемо використати готові помилки в if або створити свій тип помилки щоб гарантувати що якщо умова не виконається то далі програма викнуватись не буде

```python
class IWantCrashIt(Exception):
    def __init__(self, message="Це моя помилка, я так хочу!"):
        super().__init__(message)

if current_version not in version_restirciton:
    # raise IWantCrashIt
    pass
```

# Розказати про класи і перейти до юніт тестів

- написати клас який імітує кидання кубика (6 граней і випадання одного з чисел)

```python
import random
class Dice:
    def __init__(self, sides: int):
        self.sides = sides

    # Це методи а в нього передається завжди self - вказівка на об'єкт
    def roll(self) -> int:
        return random.randint(1, self.sides)

dice = Dice(6)
result = dice.roll()
print(f"Випало число: {result}")
```
